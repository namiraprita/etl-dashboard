```python
#sep 2022 to march 2023
selected_cols = ["Created Date",'Complaint Type','Borough']
df1 = df[selected_cols]
```

```python
df1.head(5)
```

```python
import pandas as pd


df1 = df1.assign(Created_Date=pd.to_datetime(df1['Created Date'], format="%m/%d/%Y %I:%M:%S %p"))

mask = (df1['Created_Date'] >= '2022-09-01 00:00:00') & (df1['Created_Date'] <= '2023-03-01 23:45:00')

dff = df1[mask]

df2 = dff.sort_values('Created_Date')

df2.head(20)
```

```python
selected_cols = ['Complaint Type','Borough']
df2 = df2[selected_cols]
```

```python
df2.head(20)
```

```python
df2.to_csv('/Users/zichenwang/Desktop/311finalclean.csv', index=False)
```
