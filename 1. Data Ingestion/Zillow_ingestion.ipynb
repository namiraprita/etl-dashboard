```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import base64
from IPython.display import display
from ipywidgets import interact, Dropdown
```

```python
nyc_zipcodes = pd.read_csv("nyc-zip-codes.csv")  
```

```python
# Merge rental_data with nyc_zipcodes using ZipCode/RegionName as the key
merged_df = df.merge(nyc_zipcodes, left_on='RegionName', right_on='ZipCode')
```

```python
# Delete the 'ZipCode' column on the right
merged_df.drop(columns=['ZipCode'], inplace=True)

# Check the number of NAs in the file
merged_df.isna().sum()
```

```python
# Calculate the average rental price for each zip code from 2022-01-31 to 2022-12-31
merged_df['AvgRentalPrice_2022'] = merged_df.loc[:, '2022-01-31':'2022-12-31'].mean(axis=1)
```

```python
# Recent rental price for 2023-02-28
merged_df['RecentRentalPrice'] = merged_df['2023-02-28']
```

```python
ndf = merged_df.drop(merged_df.columns[2:52],axis = 1)
```

```python
#find mean
mean = ndf['AvgRentalPrice_2022'].mean()
print(mean)
```

```python
ndf['normalized'] = ndf['AvgRentalPrice_2022']/mean
```

```python
max = ndf['normalized'].max()
min = ndf['normalized'].min()
```

```python
# Get the unique zip codes in the DataFrame
unique_zipcodes = ndf['RegionName'].unique()

# Create a dictionary to store the latitude and longitude for each zip code
zipcode_lat_lon = {}

# Set up geolocator and rate limiter
geolocator = Nominatim(user_agent="nyc_rental_prices")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

# Function to get the latitude and longitude from a zip code
def get_lat_lon(zipcode):
    location = geocode(f"{zipcode}, New York, NY")
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

# Get the latitude and longitude for each zip code and store it in the dictionary
for zipcode in unique_zipcodes:
    lat, lon = get_lat_lon(zipcode)
    zipcode_lat_lon[zipcode] = (lat, lon)

# Add latitude and longitude columns to the DataFrame
ndf['Latitude'] = ndf['RegionName'].map(lambda x: zipcode_lat_lon[x][0])
ndf['Longitude'] = ndf['RegionName'].map(lambda x: zipcode_lat_lon[x][1])
```

```python
ndf.isna().sum()
```

```python
rating_thresholds = [0.5, 0.8, 1.1, 1.4, 2.2]
rating_labels = ['A', 'B', 'C', 'D']

ndf['A_rating'] = pd.cut(ndf['normalized'], bins=rating_thresholds, labels=rating_labels)
```

```python
# Remove rows with missing latitude and longitude values
ndf = ndf.dropna(subset=['Latitude', 'Longitude'])
```

```python
ndf.to_csv('Zillowavg_final.csv')
```
